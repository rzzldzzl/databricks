# [nfusr](https://github.com/facebookarchive/nfusr) NFS Client
[nfusr: a new userspace NFS client based on libnfs - Shreyas Siravara](https://www.slideshare.net/GlusterCommunity/nfusr-a-new-userspace-nfs-client-based-on-libnfs-shreyas-siravara)

```
%sh

set -e

apt-get install -y autoconf
cd /tmp
for deb in libnfs11_2.0.0-1~exp1_amd64.deb libnfs-dev_2.0.0-1~exp1_amd64.deb; do
  wget -q "https://launchpad.net/ubuntu/+archive/primary/+files/${deb}"
  dpkg -i ${deb}
done
curl -so nfusr-master.zip https://codeload.github.com/facebookarchive/nfusr/zip/master
unzip nfusr-master.zip
cd nfusr-master
./bootstrap
./configure
make
chmod 755 /tmp/nfusr-master/nfusr /tmp/nfusr-master/mount.nfusr
make install
tar -czvf /tmp/nfusr_16.04.tgz /usr/local/bin/nfusr /sbin/mount.nfusr
mkdir -p /dbfs/nfusr
cp /tmp/nfusr_16.04.tgz /dbfs/nfusr/
cp /tmp/*.deb /dbfs/nfusr/
```

```
dbutils.fs.put("dbfs:/nfusr/nfs-client-init.sh" ,"""
dpkg -i /dbfs/nfusr/libnfs11_2.0.0-1~exp1_amd64.deb
tar -C/ -xvzf /dbfs/nfusr/nfusr_16.04.tgz
mkdir -p /mntpoint
mount.nfusr nfs://nfsserver.example.com/volume /mntpoint
""", True)
```
